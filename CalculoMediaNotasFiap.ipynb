```python
#Calculo de média dos checkpoints FIAP
# 3 CPS obrigatórios - input
# Eliminar a menor nota
# calcular média das duas maiores notas
# MCP_1sem

notas_cp = input('Insira as notas dos checkpoints (separadas por espaço):').split()
CP1 = float(notas_cp[0])
CP2 = float(notas_cp[1])
CP3 = float(notas_cp[2])

#Eliminar a menor nota - lista
notas = [CP1, CP2, CP3]
notas.sort()
MCP1 = (notas[1] + notas[2])/2

print(f'As notas inseridas foram: {CP1}, {CP2}, {CP3}')
print(f'As duas maiores notas são: {notas[1]} e {notas[2]}')
print(f'A média dos checkpoints (MCP_1sem) é: {MCP1:.2f}')
```

```python
#Calculo de média das Sprints FIAP (SP1 e SP2)
#Média das duas notas (MSP)

SP1, SP2 = input('Insira as notas das sprints:').split()

SP1 = float(SP1)
SP2 = float(SP2)

MSP1 = (SP1 + SP2)/2

print(f'A média das sprints são: {MSP1:.2f}')
```

```python
#Cálculo de notas Global Solution (GS)
#Inserir a nota
#Caso não haja nota de GS, o aluno deverá inserir sua nota de SUB

# Calcular a média parcial (CP + SP) antes da GS
md_parcial1 = 0.2 * MSP1 + 0.2 * MCP1
print(f'A média parcial (CP + SP) do primeiro semestre é: {md_parcial1:.2f}')


if (6.0 - md_parcial1) > 0:
    nota_para_passar = (6.0 - md_parcial1) / 0.6
    if nota_para_passar > 10:
        print(f'Infelizmente, mesmo com nota máxima na GS, você não alcançaria a média 6.0.')
    else:
        print(f'Para atingir 6.0 de média semestral, você precisa de no mínimo {nota_para_passar:.2f} na Global Solution.')
else:
    print('Você já tem nota suficiente de CP e SP para aprovação no primeiro semestre sem a GS.')


nota_gs1 = input('Insira a nota Global Solution (GS):')
GS1 = float(nota_gs1)

if GS1 == 0:
    nota_sub = input('Insira a nota da SUB:')
    SUB = float(nota_sub)
    GS1 = SUB # Se a SUB for inserida, ela substitui a nota da GS1
```

```python
#Cálculo da média semestral 1 (MS1)

MS1 = 0.6*GS1 + 0.2*MSP1 + 0.2*MCP1
print(f'A média semestral é: {MS1:.2f}')
```

```python
#Repetir para o segundo semestre

notas_cp2 = input('Insira as notas dos checkpoints (separadas por espaço):').split()
CP_1 = float(notas_cp2[0])
CP_2 = float(notas_cp2[1])
CP_3 = float(notas_cp2[2])

#Eliminar a menor nota - lista
notass = [CP_1, CP_2, CP_3]
notass.sort()
MCP2 = (notass[1] + notass[2])/2

print(f'As notas inseridas foram: {CP_1}, {CP_2}, {CP_3}')
print(f'As duas maiores notas são: {notass[1]} e {notass[2]}')
print(f'A média dos checkpoints (MCP_1sem) é: {MCP2:.2f}')
```

```python
SP_1, SP_2 = input('Insira as notas das sprints:').split()

SP_1 = float(SP_1)
SP_2 = float(SP_2)

MSP2 = (SP_1 + SP_2)/2

print(f'A média das sprints são: {MSP2:.2f}')
```

```python
#Cálculo de notas Global Solution (GS) para o segundo semestre

# Calcular a média parcial (CP + SP) antes da GS2
md_parcial2 = 0.2 * MSP2 + 0.2 * MCP2
print(f'A média parcial (CP + SP) do segundo semestre é: {md_parcial2:.2f}')

if (6.0 - md_parcial2) > 0:
    nota_para_passar2 = (6.0 - md_parcial2) / 0.6
    if nota_para_passar2 > 10:
        print(f'Infelizmente, mesmo com nota máxima na GS, você não alcançaria a média 6.0 no segundo semestre.')
    else:
        print(f'Para atingir 6.0 de média semestral, você precisa de no mínimo {nota_para_passar2:.2f} na Global Solution do segundo semestre.')
else:
    print('Você já tem nota suficiente de CP e SP para aprovação no segundo semestre sem a GS.')


nota_gs2 = input('Insira a nota Global Solution (GS):')
GS2 = float(nota_gs2)

if GS2 == 0:
    nota_sub2 = input('Insira a nota da SUB:')
    SUB2 = float(nota_sub2)
    GS2 = SUB2 # Se a SUB for inserida, ela substitui a nota da GS2
```

```python
#Cálculo da média semestral 2 (MS2)

MS2 = 0.6*GS2 + 0.2*MSP2 + 0.2*MCP2
print(f'A média semestral é: {MS2:.2f}')
```

```python
#Média final FIAP
# Apresentar se o aluno está aprovado >= 6.0,
#reprovado < 4.0, ou de exame >= 4.0 e <6.0 .
# Se estiver de exame, inserir nota do exame
# Calcular MF_nova = (exame + MF) /2

MF = 0.4*MS1 + 0.6*MS2
print(f'A média final é: {MF:.2f}')

if MF >= 6.0:
    print('Aprovado')
elif MF >= 4.0 and MF < 6.0:
    print('Exame')
    exame = float(input('Digite a nota do exame: '))
    MF_nova = (exame + MF)/2
    print(f'A média final é: {MF_nova:.2f}')
elif MF < 4.0:
    print('Reprovado')
```

### Função para Calcular a Média Final FIAP

Função Python que automatiza o cálculo completo da média final da FIAP, incluindo a lógica para CPs, Sprints, Global Solutions (e SUBs, se necessário) para ambos os semestres, e determina o status final do aluno (Aprovado, Exame, Reprovado).

Esta função aceitará todas as notas como argumentos, facilitando a reutilização e o teste de diferentes cenários.

```python
def calculate_fiap_final_grade(
    cp1_s1, cp2_s1, cp3_s1,
    sp1_s1, sp2_s1,
    gs1_s1_input, sub1_s1_input,
    cp1_s2, cp2_s2, cp3_s2,
    sp1_s2, sp2_s2,
    gs1_s2_input, sub1_s2_input,
    exame_input=None
):
    # --- Primeiro Semestre ---

    # Cálculo MCP1
    notas_cp_s1 = [cp1_s1, cp2_s1, cp3_s1]
    notas_cp_s1.sort()
    mcp1 = (notas_cp_s1[1] + notas_cp_s1[2]) / 2
    print(f'MCP1 (semestre 1): {mcp1:.2f}')

    # Cálculo MSP1
    msp1 = (sp1_s1 + sp2_s1) / 2
    print(f'MSP1 (semestre 1): {msp1:.2f}')

    # Cálculo GS1
    gs1 = float(gs1_s1_input)
    if gs1 == 0 and sub1_s1_input is not None:
        gs1 = float(sub1_s1_input)
    elif gs1 == 0 and sub1_s1_input is None:
        print("Aviso: GS1 é 0 e nenhuma nota SUB foi fornecida para o primeiro semestre.")
    print(f'GS1 (semestre 1): {gs1:.2f}')

    # Cálculo MS1
    ms1 = 0.6 * gs1 + 0.2 * msp1 + 0.2 * mcp1
    print(f'MS1 (Média Semestral 1): {ms1:.2f}')

    print("\n" + "-"*30 + "\n") # Separador para o segundo semestre

    # --- Segundo Semestre ---

    # Cálculo MCP2
    notas_cp_s2 = [cp1_s2, cp2_s2, cp3_s2]
    notas_cp_s2.sort()
    mcp2 = (notas_cp_s2[1] + notas_cp_s2[2]) / 2
    print(f'MCP2 (semestre 2): {mcp2:.2f}')

    # Cálculo MSP2
    msp2 = (sp1_s2 + sp2_s2) / 2
    print(f'MSP2 (semestre 2): {msp2:.2f}')

    # Cálculo GS2
    gs2 = float(gs1_s2_input) # Renomeado para gs1_s2_input para evitar conflito com GS1
    if gs2 == 0 and sub1_s2_input is not None:
        gs2 = float(sub1_s2_input)
    elif gs2 == 0 and sub1_s2_input is None:
        print("Aviso: GS2 é 0 e nenhuma nota SUB foi fornecida para o segundo semestre.")
    print(f'GS2 (semestre 2): {gs2:.2f}')

    # Cálculo MS2
    ms2 = 0.6 * gs2 + 0.2 * msp2 + 0.2 * mcp2
    print(f'MS2 (Média Semestral 2): {ms2:.2f}')

    print("\n" + "-"*30 + "\n")

    # --- Média Final ---
    mf = 0.4 * ms1 + 0.6 * ms2
    print(f'Média Final (MF): {mf:.2f}')

    final_status = ""
    if mf >= 6.0:
        final_status = 'Aprovado'
    elif mf >= 4.0 and mf < 6.0:
        print('Status: Exame')
        if exame_input is not None:
            mf_nova = (exame_input + mf) / 2
            print(f'Nota do Exame: {exame_input:.2f}')
            print(f'Média Final com Exame (MF_nova): {mf_nova:.2f}')
            if mf_nova >= 6.0:
                final_status = 'Aprovado (Após Exame)'
            else:
                final_status = 'Reprovado (Após Exame)'
        else:
            final_status = 'Exame (Nota do exame não fornecida)'
    else:
        final_status = 'Reprovado'

    print(f'Status Final: {final_status}')
    return mf, final_status
```

```python
# Exemplo de uso com as notas que você usou anteriormente:
# Primeiro Semestre:
# CPs: 10, 9, 6
# Sprints: 6, 7
# GS: 4 (sem SUB)

# Segundo Semestre:
# CPs: 8, 8, 7
# Sprints: 8, 8
# GS: 5 (sem SUB)

# Exame: 7

final_mean, status = calculate_fiap_final_grade(
    cp1_s1=10, cp2_s1=9, cp3_s1=6,
    sp1_s1=6, sp2_s1=7,
    gs1_s1_input=4, sub1_s1_input=None,
    cp1_s2=8, cp2_s2=8, cp3_s2=7,
    sp1_s2=8, sp2_s2=8,
    gs1_s2_input=5, sub1_s2_input=None,
    exame_input=7
)

```
