```python
#Calculo de média dos checkpoints FIAP
# 3 CPS obrigatórios - input
# Eliminar a menor nota
# calcular média das duas maiores notas
# MCP_1sem

notas_cp = input('Insira as notas dos checkpoints (separadas por espaço):').split()
CP1 = float(notas_cp[0])
CP2 = float(notas_cp[1])
CP3 = float(notas_cp[2])

#Eliminar a menor nota - lista
notas = [CP1, CP2, CP3]
notas.sort()
MCP1 = (notas[1] + notas[2])/2

print(f'As notas inseridas foram: {CP1}, {CP2}, {CP3}')
print(f'As duas maiores notas são: {notas[1]} e {notas[2]}')
print(f'A média dos checkpoints (MCP_1sem) é: {MCP1:.2f}')
```

```python
#Calculo de média das Sprints FIAP (SP1 e SP2)
#Média das duas notas (MSP)

SP1, SP2 = input('Insira as notas das sprints:').split()

SP1 = float(SP1)
SP2 = float(SP2)

MSP1 = (SP1 + SP2)/2

print(f'A média das sprints são: {MSP1:.2f}')
```

```python
#Cálculo de notas Global Solution (GS)
#Inserir a nota
#Caso não haja nota de GS, o aluno deverá inserir sua nota de SUB

GS1 = input('Insira a nota Global Solution (GS):')
GS1 = float(GS1)

if GS1 == 0:
    SUB = input('Insira a nota da SUB:')
```

```python
#Cálculo da média semestral 1 (MS1)

MS1 = 0.6*GS1 + 0.2*MSP1 + 0.2*MCP1
print(f'A média semestral é: {MS1:.2f}')
```

```python
#Repetir para o segundo semestre

notas_cp2 = input('Insira as notas dos checkpoints (separadas por espaço):').split()
CP_1 = float(notas_cp2[0])
CP_2 = float(notas_cp2[1])
CP_3 = float(notas_cp2[2])

#Eliminar a menor nota - lista
notass = [CP_1, CP_2, CP_3]
notass.sort()
MCP2 = (notass[1] + notass[2])/2

print(f'As notas inseridas foram: {CP_1}, {CP_2}, {CP_3}')
print(f'As duas maiores notas são: {notass[1]} e {notass[2]}')
print(f'A média dos checkpoints (MCP_1sem) é: {MCP2:.2f}')
```
